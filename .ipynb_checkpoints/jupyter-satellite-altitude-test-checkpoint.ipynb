```python
from datetime import datetime, timedelta
import math
import sys
import csv
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# example: for ISS
tlefile = './ISS-TLE.tle'
```

```python
mu = 3.986e+5

def dateconvert(timeset):
    year = 2000 + int(timeset[0:2])
    decimal_days = float(timeset[2:])
    datetimeconvert = datetime(year, 1, 1) + timedelta(decimal_days - 1)
    
    return datetimeconvert

def altitudeconvert(meanmotion):
    sma = (mu/4*(24*60*60/float(meanmotion)/math.pi)**2)**(1/3) - 6371
    
    return sma

# if you want to specify TLE file directly, please write like: 
# f = './TLE-file/ISS-test.tle'
f = open(tlefile)
line = f.readlines()
f.close

lineToList = np.array(line).reshape(-1,2).tolist()

with open('test.csv', 'w') as f:
    writer = csv.writer(f)
    for l in lineToList:
        date = dateconvert(l[0].split()[3])
        altitude = altitudeconvert(l[1].split()[7][0:11])
        writer.writerow([date, altitude])

data = pd.read_csv('test.csv', encoding = 'utf-8', header=None, names=["date", "altitude"])
data['date'] = pd.to_datetime(data['date'])
# data.dtypes

data['date'] = data['date'].dt.date
data.dtypes
```

```python
fig, ax = plt.subplots(figsize=(16,8))
plt.xticks(rotation=45)
ax.plot(data['date'], data['altitude'])

plt.show()
```
